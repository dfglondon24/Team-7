```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data representing the unemployment rates by region
data = {
    'Region': ['Western', 'Central', 'Greater Accra', 'Volta', 'Eastern', 'Ashanti', 
               'Western North', 'Ahafo', 'Bono', 'Bono East', 'Oti', 'Northern', 
               'Savannah', 'North East', 'Upper East', 'Upper West'],
    'Unemployment_Rate': ['8.8%', '7.9%', '8.2%', '7.6%', '7.0%', '7.8%', 
                          '8.1%', '7.1%', '8.0%', '6.9%', '4.2%', '7.4%', 
                          '10.5%', '9.1%', '8.9%', '6.5%']
}

# Create the DataFrame
df_unemployment = pd.DataFrame(data)

# Convert the 'Unemployment_Rate' column from percentage strings to float values
df_unemployment['Unemployment_Rate'] = df_unemployment['Unemployment_Rate'].str.replace('%', '').astype(float)

# Plot the distribution of unemployment rates by region
plt.figure(figsize=(10, 6))
df_unemployment.set_index('Region')['Unemployment_Rate'].plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('Unemployment Rates by Region')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for health insurance coverage by region
data = {
    'Region': [
        'Western', 'Central', 'Greater Accra', 'Volta', 'Eastern', 'Ashanti', 
        'Western North', 'Ahafo', 'Bono', 'Bono East', 'Oti', 'Northern', 
        'Savannah', 'North East', 'Upper East', 'Upper West'
    ],
    'Coverage_Percentage': [
        63.7, 58.0, 60.3, 67.9, 73.5, 71.9, 
        74.4, 78.1, 85.8, 78.6, 52.1, 64.8, 
        63.7, 75.4, 86.4, 83.8
    ]
}

# Create a DataFrame
df_coverage = pd.DataFrame(data)

# Plot the distribution of health insurance coverage by region
plt.figure(figsize=(12, 6))
df_coverage.set_index('Region')['Coverage_Percentage'].plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('Health Insurance Coverage Percentage by Region')
plt.xlabel('Region')
plt.ylabel('Coverage Percentage (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
